```python
def func(x) :
    try:
        return us_state_abbrev [x]
    except:
        return np.nan
    
```

```python
df['US_State'] = df['US_State'].apply(func)
```

```python
temp=[]
def func_us(row):
    
    salary=row.ConvertedCompYearlybyUSD
    us=row.US_State
    try:
        return salary / us_state_living_index[us]
    except:
        temp.append(us)
        return np.nan
    
```

```python
df['Standardization of living_us']=df.apply(func_us, axis = 1)
```
